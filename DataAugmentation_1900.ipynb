```python
#io와 pandas 모듈 import
import io
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from numpy import nan as NA
import matplotlib.pyplot as plt
```

```python
#1차 수집 데이터로 데이터 900개 만들기
#Normal
count=0
n=9
for i in range(1,101):
  filename='/content/drive/MyDrive/캡스톤디자인_인공지능/데이터수집1차/Label0_Normal/raw/Normal_case'+str(i)+'.csv'
  #pandas read_csv로 불러오기
  data= pd.read_csv(filename)
  data.columns = ['SamplingTime','AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ','-','-','-']

  data = data.drop(['-'], axis=1)

  #데이터 중에 가끔 마지막 데이터가 들어가지 않는 것이 있는데
  #그거는 삭제해준다.
  data=data.drop(data['SamplingTime'].size-1, axis=0)

  data_df=data[['AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ']]
  data_st=data[['SamplingTime']]

  #object type -> float 타입으로 변환시켜줌
  data_df[['AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ']] = data_df[['AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ']].apply(pd.to_numeric)

  #--------------------------데이터 정리------------------------------

  #데이터에 더해줄 값을 정규분포에서 랜덤하게 번호 뽑기
  #9개를 뽑아서 파일 한 개당 파일 9개를 생성하면 총 900개 생성됨
  acX_rand=[]
  acY_rand=[]
  acZ_rand=[]
  gyroX_rand=[]
  gyroY_rand=[]
  gyroZ_rand=[]

  #데이터에 더해줄 값을 정규분포에서 랜덤하게 번호 뽑기
  for x in range(n):
    acX_rand.append(np.random.normal(data_df['AcX'].mean(), data_df['AcX'].std()*0.7, size=len(data_df['AcX'])))
    acY_rand.append(np.random.normal(data_df['AcY'].mean(), data_df['AcY'].std()*0.7, size=len(data_df['AcY'])))
    acZ_rand.append(np.random.normal(data_df['AcZ'].mean(), data_df['AcZ'].std()*0.7, size=len(data_df['AcZ'])))
    gyroX_rand.append(np.random.normal(data_df['GyroX'].mean(), data_df['GyroX'].std()*0.7, size=len(data_df['GyroX'])))
    gyroY_rand.append(np.random.normal(data_df['GyroY'].mean(), data_df['GyroY'].std()*0.7, size=len(data_df['GyroY'])))
    gyroZ_rand.append(np.random.normal(data_df['GyroZ'].mean(), data_df['GyroZ'].std()*0.7, size=len(data_df['GyroZ'])))
  #shape : (9,len(data_df['AcX']))
  acX_rand=np.array(acX_rand) 
  acY_rand=np.array(acY_rand)
  acZ_rand=np.array(acZ_rand)
  gyroX_rand=np.array(gyroX_rand)
  gyroY_rand=np.array(gyroY_rand)
  gyroZ_rand=np.array(gyroZ_rand)
  #shape : ->(len(data_df['AcX']), 9)로 변경
  acX_rand=acX_rand.transpose() 
  acY_rand=acY_rand.transpose()
  acZ_rand=acZ_rand.transpose()
  gyroX_rand=gyroX_rand.transpose()
  gyroY_rand=gyroY_rand.transpose()
  gyroZ_rand=gyroZ_rand.transpose()

  new_acX = []
  new_acY = []
  new_acZ = []
  new_gyroX = []
  new_gyroY = []
  new_gyroZ = []
  #위에 구한 랜덤한 수를 더해서 새로운 데이터 생성
  for y in range(n):
    for z in range(len(data_df['AcX'])):
      new_acX.append([data_df['AcX'][z]+acX_rand[z][y]])
      new_acY.append([data_df['AcY'][z]+acY_rand[z][y]])
      new_acZ.append([data_df['AcZ'][z]+acZ_rand[z][y]])
      new_gyroX.append([data_df['GyroX'][z]+gyroX_rand[z][y]])
      new_gyroY.append([data_df['GyroY'][z]+gyroY_rand[z][y]])
      new_gyroZ.append([data_df['GyroZ'][z]+gyroZ_rand[z][y]])
  #shape : (len(data_df['AcX'])*9, 1)
  new_acX=np.array(new_acX)
  new_acY=np.array(new_acY)
  new_acZ=np.array(new_acZ)
  new_gyroX=np.array(new_gyroX)
  new_gyroY=np.array(new_gyroY)
  new_gyroZ=np.array(new_gyroZ)
  #shape : (len(data_df['AcX])*9, 6)
  new_total=np.concatenate((new_acX,new_acY,new_acZ,new_gyroX,new_gyroY,new_gyroZ), axis=1)
  #마지막으로 데이터 파일 생성
  total=[]
  for j in range(n):
    new_total=pd.DataFrame(new_total)
    total=new_total.loc[j*len(data_df['AcX']):(j+1)*(len(data_df['AcX'])-1)]
    total.columns = ['AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ']
    count += 1
    total.to_csv('/content/drive/MyDrive/캡스톤디자인_인공지능/데이터생성/Label0_Normal/raw/Normal_case'+str(count)+'.csv')
```
