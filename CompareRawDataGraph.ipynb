```python
filename_label0='/content/drive/MyDrive/캡스톤디자인_인공지능/데이터수집1차/Label0_Normal/raw/Normal_case10.csv'
filename_label1='/content/drive/MyDrive/캡스톤디자인_인공지능/데이터수집1차/Label1_warning1/raw/Warning1_case10.csv'
filename_label2='/content/drive/MyDrive/캡스톤디자인_인공지능/데이터수집1차/Label2_warning2/raw/CASE3_10.csv'
filename_label3='/content/drive/MyDrive/캡스톤디자인_인공지능/데이터수집1차/Label3_warning3/raw/CASE4_10.csv'
```

```python
data_label0=pd.read_csv(filename_label0)
data_label0.columns = ['SamplingTime','AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ','-','-','-']

data_label1=pd.read_csv(filename_label1)
data_label1.columns = ['SamplingTime','AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ','-','-','-']

data_label2=pd.read_csv(filename_label2)
data_label2.columns = ['SamplingTime','AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ','-','-','-']

data_label3=pd.read_csv(filename_label3)
data_label3.columns = ['SamplingTime','AcX', 'AcY','AcZ','GyroX','GyroY','GyroZ','-','-','-']
```
